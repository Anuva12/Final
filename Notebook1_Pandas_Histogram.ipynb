This dataset doesn't include the attribute names in its header so we need to specify them manually, or the first line of the file will be regarded as the header.

```python
iris = pd.read_csv('./iris/iris.data', names=['sepal_length', 'sepal_width', 'petal_length', 'petal_width', 'class'])
```

```python
iris.sample(10)
```

```python
plt.subplot(1, 2, 1)
plt.hist(iris['sepal_length'], bins=50)
plt.xlabel('Sepal Length')
plt.ylabel('Count')
plt.title('Histogram of Sepal length with 50 bins')
plt.subplots_adjust(bottom=0, top=1, left=-0.5, right=1.5)
plt.subplot(1, 2, 2)
plt.hist(iris['sepal_length'], bins=20)
plt.xlabel('Sepal Length')
plt.ylabel('Count')
plt.title('Histogram of Sepal length with 20 bins')
plt.subplots_adjust(bottom=0, top=1, left=-0.5, right=1.5)
```

```python
n_bins = 20

setosa = iris[iris['class'] == 'Iris-setosa']
versicolor = iris[iris['class'] == 'Iris-versicolor']
virginica = iris[iris['class'] == 'Iris-virginica']


colors = ['orange', 'blue', 'black']
labels = ['setosa','versicolor','virginica']
x_multi = [setosa['sepal_length'],versicolor['sepal_length'],virginica['sepal_length']]
plt.hist(x_multi, n_bins, histtype='bar',color=colors,label= labels)
plt.legend(prop={'size': 10})
plt.xlabel('Sepal Length')
plt.ylabel('Count')
```

```python
plants = pd.read_csv('nuclear-plants.txt', sep = "\t")

# A: outliers in Cost:

def outliers(paramStr):
  IQR = plants[paramStr].quantile(0.75) - plants[paramStr].quantile(0.25)
  outliers = plants[(plants[paramStr] < (plants[paramStr].quantile(0.25) - 1.5 * IQR)) |
    (plants[paramStr] > (plants[paramStr].quantile(0.75) + 1.5 * IQR))]
  print("Outliers in ", paramStr, ": ", outliers)

outliers("Cost")
outliers("Mwatts")
outliers("Date")
print()

# B: Power Plant Data
meanCost = plants["Cost"].mean()
stdCost = plants["Cost"].std()
print("POWER PLANTS: Mean Cost: ", meanCost, " , Standard Deviation: ", stdCost)
print()

# C: Cost per Megawatt
plants["Cost_MW"] = plants["Cost"] / plants["Mwatts"]
meanCpmw = plants["Cost_MW"].mean()
stdCpmw = plants["Cost_MW"].std()
print("COST PER MEGAWATT: Mean Cost: ", meanCpmw, " , Standard Deviation: ", stdCpmw)
print()

# D: Histogram
n_bins = 15
plt.hist(plants["Cost_MW"], n_bins, histtype='bar')
plt.xlabel('Cost Per Megawatt')
plt.ylabel('Count')
plt.title("Histogram of Cost Per Megawatt")
plt.show()
print()

skewness = plants["Cost_MW"].skew()
print("The skewness is: ", skewness)
print('''The plot is very sligtly right skewed, indicating that the cost per megawatt is
  usually lower but overall very equally distributed. ''')
print()

# plants.head()
```

```python
hotDogs = pd.read_csv('hot-dogs-1.txt', sep = "\t")\

n_bins = 20

beef = hotDogs[hotDogs['Type'] == 'Beef']
poultry = hotDogs[hotDogs['Type'] == 'Poultry']
meat = hotDogs[hotDogs['Type'] == 'Meat']

colors = ['orange', 'blue', 'black']
labels = ['Beef', 'Poultry', 'Meat']

# Sodium Content histogram
x_multi_sodium = [beef['Sodium'], poultry['Sodium'], meat['Sodium']]
plt.hist(x_multi_sodium, n_bins, histtype='bar', color=colors, label=labels)
plt.legend(prop={'size': 10})
plt.xlabel('Sodium (mg)')
plt.ylabel('Count')
plt.show()

# Calories histogram
x_multi_calories = [beef['Calories'], poultry['Calories'], meat['Calories']]
plt.hist(x_multi_calories, n_bins, histtype='bar', color=colors, label=labels)
plt.legend(prop={'size': 10})
plt.xlabel('Calories')
plt.ylabel('Count')
plt.show()

# hotDogs.head()
```

```python
reads = pd.read_csv('Readability.csv', sep = ",")

reads.head()

# (a)
reads.boxplot(column='3SYL', by='GROUP')
plt.title('Number of 3+ Syllable Words by Reader Group')
plt.xlabel('Reader Group (GRP)')
plt.ylabel('Number of 3+ Syllable Words')
plt.show()

# (b)
plt.figure(figsize=(8,6))
reads.boxplot(column='SEN', by='GROUP')
plt.title('Number of Sentences by Reader Group')
plt.xlabel('Reader Group (GRP)')
plt.ylabel('Number of Sentences')
plt.show()

```
